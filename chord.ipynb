```python
cm = np.array([
    [829, 335, 13, 15, 175, 79],
    [254, 889, 60, 27, 303, 71],
    [72, 241, 109, 18, 339, 10],
    [100, 236, 38, 59, 488, 55],
    [175, 272, 73, 69, 1382, 75],
    [135, 91, 2, 19, 99, 186],
])
print(cm.shape)
total = np.sum(cm, axis=0).sum()
print(total)
print("Percent class predicted")
print(np.sum(cm, axis=0) / total)
print("Percent class actual")
print(np.sum(cm, axis=1) / total)
assert np.sum(cm, axis=0).sum() == np.sum(cm, axis=1).sum() 
```

```python
links = []
for i in range(len(cm)):
    for j in range(len(cm)):
        if i == j:
            continue
        links.append([i, j, cm[i][j]])

# Create the dataframe
links_df = pd.DataFrame(links, columns=['source', 'target', 'value'])
nodes = ["Stop-G1", "G1", "G1-S", "S-G2", "G2", "G2-M"]
nodes_df = pd.DataFrame(nodes, columns=['cluster'])
nodes_ds = hv.Dataset(pd.DataFrame(nodes_df['nodes']), 'index')


chord = hv.Chord((links_df, nodes_ds)).select(value=(5, None))
chord.opts(
    opts.Chord(cmap='Category20', edge_cmap='Category20', edge_color=dim('source').str(), 
               labels='name', node_color=dim('index').str())
)
```
